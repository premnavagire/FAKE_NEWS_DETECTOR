```python
# ══════════════════════════════════════════════
# FINAL CLUSTERING CODE (FULLY CORRECTED)
# ══════════════════════════════════════════════

!pip install -q umap-learn

# Imports (fixed missing ones)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import TruncatedSVD, PCA
from sklearn.metrics import silhouette_score, adjusted_rand_score, normalized_mutual_info_score

# ══════════════════════════════════════════════
# LOAD YOUR DATASET HERE (ADAPT TO YOUR FILE)
# ══════════════════════════════════════════════
# Assuming df_all has at least two columns: 'title' (str) and 'label' (0/1)
# Example: df_all = pd.read_csv('your_news_data.csv')

# For demonstration, let's create a dummy dataset (remove this in production)
# You MUST replace this with your actual data loading code.
from sklearn.datasets import fetch_20newsgroups
news = fetch_20newsgroups(subset='all', categories=['rec.sport.baseball', 'sci.space'], shuffle=True, random_state=42)
df_all = pd.DataFrame({'title': news.data[:1000], 'label': news.target[:1000]})  # dummy labels

# Create clean_title (lowercase, remove punctuation)
import re
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text

df_all['clean_title'] = df_all['title'].apply(clean_text)
print("✅ Data loaded and 'clean_title' created")

# ══════════════════════════════════════════════
# STEP 1: TF-IDF
# ══════════════════════════════════════════════
tfidf = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf.fit_transform(df_all['clean_title'])
print("✅ TF-IDF created successfully!")

# ══════════════════════════════════════════════
# STEP 2: SVD (Dimension Reduction)
# ══════════════════════════════════════════════
svd = TruncatedSVD(n_components=100, random_state=42)
X_svd = svd.fit_transform(X_tfidf)
print(f"SVD explained variance: {svd.explained_variance_ratio_.sum():.2%}")

# ══════════════════════════════════════════════
# STEP 3: FIND BEST K (2 to 5)
# ══════════════════════════════════════════════
inertias = []
sil_scores = []
K_range = range(2, 6)

print("\nFinding best K...")
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_svd)
    inertias.append(km.inertia_)
    sil = silhouette_score(X_svd, labels)
    sil_scores.append(sil)
    print(f"K={k} | Inertia={km.inertia_:.0f} | Silhouette={sil:.4f}")

# Plot elbow & silhouette
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(K_range, inertias, 'o-')
plt.title("Elbow Method")
plt.subplot(1,2,2)
plt.plot(K_range, sil_scores, 'o-', color='green')
plt.title("Silhouette Score")
plt.tight_layout()
plt.show()

best_k = K_range[sil_scores.index(max(sil_scores))]
print(f"\n✅ Best K = {best_k}")

# ══════════════════════════════════════════════
# STEP 4: FINAL KMEANS
# ══════════════════════════════════════════════
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df_all['cluster'] = kmeans.fit_predict(X_svd)

print("\nCluster Distribution:")
print(df_all['cluster'].value_counts())

# ══════════════════════════════════════════════
# STEP 5: FAKE % PER CLUSTER
# ══════════════════════════════════════════════
cluster_stats = df_all.groupby('cluster')['label'].mean() * 100
print("\nFake % per cluster:")
print(cluster_stats)

plt.figure(figsize=(6,4))
cluster_stats.plot(kind='bar', color=['green','red'])
plt.title("Fake % per Cluster")
plt.ylabel("Percentage")
plt.show()

# ══════════════════════════════════════════════
# STEP 6: PCA VISUALIZATION
# ══════════════════════════════════════════════
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_svd)

plt.figure(figsize=(8,6))
scatter = plt.scatter(X_2d[:,0], X_2d[:,1], c=df_all['cluster'], cmap='viridis', alpha=0.5)
plt.colorbar(scatter, label='Cluster')
plt.title("Clusters Visualization (PCA)")
plt.show()

# ══════════════════════════════════════════════
# STEP 7: HEATMAP (Cluster vs Real/Fake)
# ══════════════════════════════════════════════
pivot = pd.crosstab(df_all['cluster'], df_all['label'])
pivot.columns = ['Real', 'Fake']  # assuming label 0=real, 1=fake
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlOrRd')
plt.title("Cluster vs Real/Fake")
plt.show()

# ══════════════════════════════════════════════
# STEP 8: DBSCAN
# ══════════════════════════════════════════════
db = DBSCAN(eps=1.5, min_samples=10)
df_all['dbscan'] = db.fit_predict(X_svd)
print("\nDBSCAN clusters:", set(df_all['dbscan']))
print("Outliers (cluster -1):", (df_all['dbscan']==-1).sum())

# ══════════════════════════════════════════════
# STEP 9: METRICS
# ══════════════════════════════════════════════
sil = silhouette_score(X_svd, df_all['cluster'])
ari = adjusted_rand_score(df_all['label'], df_all['cluster'])
nmi = normalized_mutual_info_score(df_all['label'], df_all['cluster'])

print("\n📊 Clustering Metrics:")
print(f"Silhouette Score : {sil:.4f}")
print(f"Adjusted Rand    : {ari:.4f}")
print(f"NMI Score        : {nmi:.4f}")

print("\n✅ CLUSTERING DONE 🚀")
```

```python
# ══════════════════════════════════════════════
# PHASE 4: BASELINE ML CLASSIFICATION (CORRECTED)
# Why: Establish performance benchmarks
# Models: Logistic Regression + XGBoost
# Metric: PR-AUC (handles imbalance correctly)
# ══════════════════════════════════════════════

!pip install -q xgboost imbalanced-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (classification_report,
                              roc_auc_score,
                              precision_recall_curve,
                              auc, confusion_matrix, f1_score)
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE

# ══════════════════════════════════════════════
# LOAD YOUR REAL DATA HERE (REPLACE THIS SECTION)
# ══════════════════════════════════════════════
# Expected: X_combined = feature matrix (n_samples, n_features)
#           y = label array (0=real, 1=fake)
#
# Example if you have a DataFrame 'df_all' with 'clean_title' and 'label':
#
# from sklearn.feature_extraction.text import TfidfVectorizer
# tfidf = TfidfVectorizer(max_features=5000)
# X_combined = tfidf.fit_transform(df_all['clean_title']).toarray()  # or keep sparse
# y = df_all['label'].values
#
# For this demo, we create dummy data (remove when using real data)
np.random.seed(42)
n_samples = 5000
n_features = 100
X_combined = np.random.randn(n_samples, n_features)
y = np.random.binomial(1, 0.3, size=n_samples)  # 30% fake
print("⚠️ Using dummy data. Replace with your actual X_combined and y.")
print(f"X_combined shape: {X_combined.shape}, y shape: {y.shape}")

# ── 4A. Train/Test Split ──
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y,
    test_size=0.2,
    random_state=42,
    stratify=y       # keeps same fake/real ratio in both splits
)
print(f"Train: {X_train.shape} | Test: {X_test.shape}")
print(f"Train labels: {dict(pd.Series(y_train).value_counts())}")

# ── 4B. Handle imbalance with SMOTE ──
# Why SMOTE: creates synthetic fake samples so model learns fake news properly
sm = SMOTE(random_state=42)
X_train_bal, y_train_bal = sm.fit_resample(X_train, y_train)
print(f"\nAfter SMOTE: {dict(pd.Series(y_train_bal).value_counts())}")

# ── 4C. Evaluation helper function ──
def evaluate(name, clf, X_tr, y_tr, X_te, y_te):
    clf.fit(X_tr, y_tr)
    y_pred  = clf.predict(X_te)
    y_proba = clf.predict_proba(X_te)[:, 1]

    roc  = roc_auc_score(y_te, y_proba)
    p, r, _ = precision_recall_curve(y_te, y_proba)
    pr   = auc(r, p)
    f1   = f1_score(y_te, y_pred)

    print(f"\n{'='*50}")
    print(f"Model : {name}")
    print(f"ROC-AUC : {roc:.4f}  |  PR-AUC : {pr:.4f}  |  F1 : {f1:.4f}")
    print(classification_report(y_te, y_pred,
                                 target_names=['Real','Fake']))

    # PR curve plot
    plt.figure(figsize=(6, 4))
    plt.plot(r, p, color='#534AB7', lw=2, label=f'PR-AUC={pr:.3f}')
    plt.xlabel('Recall'); plt.ylabel('Precision')
    plt.title(f'PR Curve — {name}')
    plt.legend(); plt.grid(alpha=0.3)
    plt.tight_layout(); plt.show()

    return clf, y_proba, pr

# ── 4D. Logistic Regression ──
lr = LogisticRegression(C=1.0, max_iter=1000,
                         solver='lbfgs', random_state=42)
lr_clf, lr_proba, lr_pr = evaluate(
    "Logistic Regression",
    lr, X_train_bal, y_train_bal, X_test, y_test
)

# ── 4E. XGBoost ──
# Compute positive class weight (for imbalance handling)
pos_weight = (y_train == 0).sum() / (y_train == 1).sum()  # real/fake ratio
xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    scale_pos_weight=pos_weight,
    use_label_encoder=False,       # suppress warning
    eval_metric='logloss',
    random_state=42,
    n_jobs=-1
)
xgb_clf, xgb_proba, xgb_pr = evaluate(
    "XGBoost",
    xgb, X_train, y_train, X_test, y_test
)

# ── 4F. Compare both models ──
# Overlaid PR curves
plt.figure(figsize=(8, 6))
for proba, name, color in [
    (lr_proba,  'Logistic Regression', '#534AB7'),
    (xgb_proba, 'XGBoost',             '#D85A30')]:
    p, r, _ = precision_recall_curve(y_test, proba)
    pr = auc(r, p)
    plt.plot(r, p, lw=2, color=color, label=f'{name} (PR={pr:.3f})')

plt.xlabel('Recall'); plt.ylabel('Precision')
plt.title('PR Curves — All Models Compared')
plt.legend(); plt.grid(alpha=0.3)
plt.tight_layout(); plt.show()

# Confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for ax, proba, name in [
    (axes[0], lr_proba,  'Logistic Regression'),
    (axes[1], xgb_proba, 'XGBoost')]:
    cm = confusion_matrix(y_test, (proba >= 0.5).astype(int))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                xticklabels=['Real','Fake'],
                yticklabels=['Real','Fake'])
    ax.set_title(f'{name}')
plt.tight_layout(); plt.show()

print("\n✓ Phase 4 complete!")
```

```python
# ══════════════════════════════════════════════
# CELL 3: WORD EMBEDDING
# Paper Section: WORD EMBEDDING
# Methods: Word2Vec, GloVe (as mentioned in paper)
# ══════════════════════════════════════════════

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Parameters
VOCAB_SIZE  = 10000
MAX_LEN     = 50      # Paper: avg title = 17.9 words
EMBED_DIM   = 100     # Paper uses GloVe 100-dim

texts  = df_all['clean_title'].tolist()
labels = df_all['label'].values

# ── Tokenization ──
# Paper: preprocessing includes tokenization
tokenizer_keras = Tokenizer(
    num_words=VOCAB_SIZE,
    oov_token='<OOV>'
)
tokenizer_keras.fit_on_texts(texts)
sequences = tokenizer_keras.texts_to_sequences(texts)
X_padded  = pad_sequences(sequences,
                            maxlen=MAX_LEN,
                            padding='post',
                            truncating='post')
word_index = tokenizer_keras.word_index
print(f"Vocabulary size: {len(word_index)}")
print(f"X_padded shape : {X_padded.shape}")

# ── Train/Test Split ──
# Paper: 80% train, 20% test
X_train,X_test,y_train,y_test = train_test_split(
    X_padded, labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)
print(f"\nTrain: {X_train.shape}")
print(f"Test : {X_test.shape}")
print(f"Train labels: {dict(pd.Series(y_train).value_counts())}")

# ── Download GloVe ──
# Paper [17] uses GloVe 100-dim embeddings
print("\n⏳ Downloading GloVe embeddings...")
import urllib.request
glove_url = ("http://nlp.stanford.edu/data/"
             "glove.6B.zip")

# Use smaller pre-downloaded version
!wget -q "https://downloads.cs.stanford.edu/nlp/data/glove.6B.zip"
!unzip -q glove.6B.zip -d glove/
print("✓ GloVe downloaded!")

# Load GloVe
glove_embeddings = {}
with open('/content/glove/glove.6B.100d.txt',
          'r', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word   = values[0]
        vector = np.asarray(values[1:], dtype='float32')
        glove_embeddings[word] = vector

print(f"✓ GloVe loaded: {len(glove_embeddings)} words")

# Create embedding matrix
embedding_matrix = np.zeros(
    (VOCAB_SIZE, EMBED_DIM))
found = 0
for word, idx in word_index.items():
    if idx < VOCAB_SIZE:
        vec = glove_embeddings.get(word)
        if vec is not None:
            embedding_matrix[idx] = vec
            found += 1

print(f"✓ Words found in GloVe: {found}/{VOCAB_SIZE}")
print("\n✅ CELL 3 done!")
```

```python
# ══════════════════════════════════════════════
# CELL 4: CNN MODEL
# Paper [17]: FNDNet — Deep CNN
# Architecture: GloVe → Conv1D → MaxPool → Dense
# Accuracy: 98.36% on Kaggle dataset
# ══════════════════════════════════════════════

from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Embedding, Conv1D, MaxPooling1D,
    GlobalMaxPooling1D, Dense, Dropout,
    Flatten, concatenate, BatchNormalization)
from tensorflow.keras.callbacks import (
    EarlyStopping, ModelCheckpoint)

def build_cnn_model():
    """
    FNDNet architecture from Paper [17]
    3 parallel Conv1D with different kernel sizes
    Captures different n-gram patterns
    """
    inp = Input(shape=(MAX_LEN,))

    # GloVe embedding layer
    emb = Embedding(
        input_dim=VOCAB_SIZE,
        output_dim=EMBED_DIM,
        weights=[embedding_matrix],
        input_length=MAX_LEN,
        trainable=False    # freeze GloVe weights
    )(inp)

    # 3 parallel Conv1D (different kernel sizes)
    # Paper: different kernel sizes capture
    # different patterns
    conv1 = Conv1D(128, 2,
                   activation='relu',
                   padding='same')(emb)
    pool1 = MaxPooling1D(2)(conv1)

    conv2 = Conv1D(128, 3,
                   activation='relu',
                   padding='same')(emb)
    pool2 = MaxPooling1D(2)(conv2)

    conv3 = Conv1D(128, 4,
                   activation='relu',
                   padding='same')(emb)
    pool3 = MaxPooling1D(2)(conv3)

    # Merge parallel convolutions
    merged = concatenate([
        GlobalMaxPooling1D()(pool1),
        GlobalMaxPooling1D()(pool2),
        GlobalMaxPooling1D()(pool3)
    ])

    # Dense layers (fully connected)
    x = Dense(128, activation='relu')(merged)
    x = Dropout(0.5)(x)
    x = BatchNormalization()(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.3)(x)

    # Output (sigmoid for binary classification)
    out = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inp, outputs=out)
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

cnn_model = build_cnn_model()
cnn_model.summary()

# Train
print("\n⏳ Training CNN Model...")
history_cnn = cnn_model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=10,
    batch_size=64,
    callbacks=[
        EarlyStopping(patience=3,
                       restore_best_weights=True),
        ModelCheckpoint('/content/cnn_best.h5',
                         save_best_only=True)
    ],
    verbose=1
)

# Evaluate
cnn_proba = cnn_model.predict(X_test).flatten()
cnn_pred  = (cnn_proba >= 0.5).astype(int)
cnn_f1    = f1_score(y_test, cnn_pred)
p,r,_     = precision_recall_curve(y_test,cnn_proba)
cnn_pr    = auc(r,p)
cnn_roc   = roc_auc_score(y_test,cnn_proba)

print(f"\n📊 CNN Results:")
print(f"  ROC-AUC : {cnn_roc:.4f}")
print(f"  PR-AUC  : {cnn_pr:.4f}")
print(f"  F1      : {cnn_f1:.4f}")
print(classification_report(y_test,cnn_pred,
      target_names=['Real','Fake']))

# Plot training history
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,4))
ax1.plot(history_cnn.history['accuracy'],
         label='Train')
ax1.plot(history_cnn.history['val_accuracy'],
         label='Val')
ax1.set_title('CNN Accuracy'); ax1.legend()
ax2.plot(history_cnn.history['loss'],
         label='Train')
ax2.plot(history_cnn.history['val_loss'],
         label='Val')
ax2.set_title('CNN Loss'); ax2.legend()
plt.tight_layout(); plt.show()

print("\n✅ CELL 4 CNN done!")
```

```python
# ══════════════════════════════════════════════
# CELL 5: -LSTM MODEL
# Paper Section: Recurrent Neural Networks
# Paper [19]: CNN +-LSTM → 92% accuracy
# LSTM: handles sequential text patterns
# ══════════════════════════════════════════════

from tensorflow.keras.layers import (
    LSTM, Bidirectional, GRU)

def build_lstm_model():
    """
    BiLSTM model
    Paper: LSTM captures long-term dependencies
    BiLSTM reads text both forward and backward
    """
    inp = Input(shape=(MAX_LEN,))

    emb = Embedding(
        input_dim=VOCAB_SIZE,
        output_dim=EMBED_DIM,
        weights=[embedding_matrix],
        input_length=MAX_LEN,
        trainable=False
    )(inp)

    # Bidirectional LSTM
    # Paper [4]: BiLSTM used for fake news
    x = Bidirectional(
        LSTM(64, return_sequences=True)
    )(emb)
    x = Dropout(0.3)(x)

    x = Bidirectional(
        LSTM(32, return_sequences=False)
    )(x)
    x = Dropout(0.3)(x)

    x = Dense(64, activation='relu')(x)
    x = Dropout(0.3)(x)
    out = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inp, outputs=out)
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

lstm_model = build_lstm_model()
lstm_model.summary()

print("\n⏳ Training LSTM Model...")
history_lstm = lstm_model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=10,
    batch_size=64,
    callbacks=[
        EarlyStopping(patience=3,
                       restore_best_weights=True),
        ModelCheckpoint('/content/lstm_best.h5',
                         save_best_only=True)
    ],
    verbose=1
)

# Evaluate
lstm_proba = lstm_model.predict(X_test).flatten()
lstm_pred  = (lstm_proba >= 0.5).astype(int)
lstm_f1    = f1_score(y_test, lstm_pred)
p,r,_      = precision_recall_curve(y_test,lstm_proba)
lstm_pr    = auc(r,p)
lstm_roc   = roc_auc_score(y_test,lstm_proba)

print(f"\n📊 LSTM Results:")
print(f"  ROC-AUC : {lstm_roc:.4f}")
print(f"  PR-AUC  : {lstm_pr:.4f}")
print(f"  F1      : {lstm_f1:.4f}")
print(classification_report(y_test,lstm_pred,
      target_names=['Real','Fake']))

print("\n✅ CELL 5 LSTM done!")
```

```python
# ══════════════════════════════════════════════
# CELL 6: HYBRID CNN + LSTM
# Paper [19]: CNN + RNN-LSTM → 92% accuracy
# CNN extracts features → LSTM learns sequences
# Best of both worlds!
# ══════════════════════════════════════════════

def build_hybrid_model():
    """
    Hybrid CNN + BiLSTM
    Paper [19]: hybrid achieved 92%
    CNN: 90%, RNN: 90%
    Hybrid > both individually!
    """
    inp = Input(shape=(MAX_LEN,))

    emb = Embedding(
        input_dim=VOCAB_SIZE,
        output_dim=EMBED_DIM,
        weights=[embedding_matrix],
        input_length=MAX_LEN,
        trainable=False
    )(inp)

    # CNN part — extract high-level features
    conv = Conv1D(128, 3,
                  activation='relu',
                  padding='same')(emb)
    conv = BatchNormalization()(conv)
    pool = MaxPooling1D(2)(conv)
    drop = Dropout(0.3)(pool)

    # LSTM part — learn sequential patterns
    lstm = Bidirectional(
        LSTM(64, return_sequences=False)
    )(drop)
    drop2 = Dropout(0.3)(lstm)

    # Dense layers
    x   = Dense(64, activation='relu')(drop2)
    x   = Dropout(0.3)(x)
    out = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inp, outputs=out)
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

hybrid_model = build_hybrid_model()
hybrid_model.summary()

print("\n⏳ Training Hybrid CNN+LSTM...")
history_hyb = hybrid_model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=10,
    batch_size=64,
    callbacks=[
        EarlyStopping(patience=3,
                       restore_best_weights=True),
        ModelCheckpoint('/content/hybrid_best.h5',
                         save_best_only=True)
    ],
    verbose=1
)

# Evaluate
hyb_proba = hybrid_model.predict(X_test).flatten()
hyb_pred  = (hyb_proba >= 0.5).astype(int)
hyb_f1    = f1_score(y_test, hyb_pred)
p,r,_     = precision_recall_curve(y_test,hyb_proba)
hyb_pr    = auc(r,p)
hyb_roc   = roc_auc_score(y_test,hyb_proba)

print(f"\n📊 Hybrid CNN+LSTM Results:")
print(f"  ROC-AUC : {hyb_roc:.4f}")
print(f"  PR-AUC  : {hyb_pr:.4f}")
print(f"  F1      : {hyb_f1:.4f}")
print(classification_report(y_test,hyb_pred,
      target_names=['Real','Fake']))

print("\n✅ CELL 6 Hybrid done!")
```

```python
# ══════════════════════════════════════════════
# CELL 7: ATTENTION MODEL + BERT
# Paper Section: Attention Mechanism
# Paper [31]: Attention-based C-BiLSTM
# Paper [15]: CMS with Self-attention
# BERT: context-based model (paper section)
# ══════════════════════════════════════════════

from tensorflow.keras.layers import (
    MultiHeadAttention, LayerNormalization,
    GlobalAveragePooling1D)

# ── Part A: Attention-based Model ──
def build_attention_model():
    """
    CNN + BiLSTM + Attention
    Paper [31]: AC-BiLSTM
    Attention updates weights for important words
    """
    inp = Input(shape=(MAX_LEN,))

    emb = Embedding(
        input_dim=VOCAB_SIZE,
        output_dim=EMBED_DIM,
        weights=[embedding_matrix],
        input_length=MAX_LEN,
        trainable=False
    )(inp)

    # CNN feature extraction
    conv = Conv1D(64, 3,
                  activation='relu',
                  padding='same')(emb)
    conv = MaxPooling1D(2)(conv)

    # BiLSTM
    lstm_out = Bidirectional(
        LSTM(64, return_sequences=True)
    )(conv)

    # Multi-head Self-Attention
    # Paper [15]: CMS uses self-attention
    attn = MultiHeadAttention(
        num_heads=4, key_dim=32
    )(lstm_out, lstm_out)
    attn = LayerNormalization()(attn + lstm_out)

    # Global pooling
    x   = GlobalAveragePooling1D()(attn)
    x   = Dense(64, activation='relu')(x)
    x   = Dropout(0.3)(x)
    out = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inp, outputs=out)
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

attn_model = build_attention_model()
attn_model.summary()

print("⏳ Training Attention Model...")
history_attn = attn_model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=10,
    batch_size=64,
    callbacks=[
        EarlyStopping(patience=3,
                       restore_best_weights=True)
    ],
    verbose=1
)

# Evaluate
attn_proba = attn_model.predict(X_test).flatten()
attn_pred  = (attn_proba >= 0.5).astype(int)
attn_f1    = f1_score(y_test, attn_pred)
p,r,_      = precision_recall_curve(y_test,attn_proba)
attn_pr    = auc(r,p)
attn_roc   = roc_auc_score(y_test,attn_proba)

print(f"\n📊 Attention Model Results:")
print(f"  ROC-AUC : {attn_roc:.4f}")
print(f"  PR-AUC  : {attn_pr:.4f}")
print(f"  F1      : {attn_f1:.4f}")

# ── Part B: BERT ──
print("\n⏳ Loading BERT (DistilBERT)...")
!pip install -q transformers datasets

import torch
from transformers import (AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments, Trainer)
from datasets import Dataset
from sklearn.metrics import precision_recall_curve,auc,f1_score

bert_tokenizer = AutoTokenizer.from_pretrained(
    'distilbert-base-uncased')

df_bert = df_all[['clean_title','label']].copy()
df_bert.columns = ['title','label']
tr_df,te_df = train_test_split(
    df_bert, test_size=0.2,
    stratify=df_bert['label'], random_state=42)

train_ds = Dataset.from_pandas(
    tr_df.reset_index(drop=True))
test_ds  = Dataset.from_pandas(
    te_df.reset_index(drop=True))

def tokenize_bert(batch):
    return bert_tokenizer(batch['title'],
        padding='max_length',
        truncation=True, max_length=64)

train_ds = train_ds.map(tokenize_bert,batched=True)
test_ds  = test_ds.map(tokenize_bert,batched=True)
train_ds = train_ds.rename_column('label','labels')
test_ds  = test_ds.rename_column('label','labels')
train_ds.set_format('torch',
    columns=['input_ids','attention_mask','labels'])
test_ds.set_format('torch',
    columns=['input_ids','attention_mask','labels'])

bert_model = AutoModelForSequenceClassification\
    .from_pretrained('distilbert-base-uncased',
                     num_labels=2)

def compute_metrics(eval_pred):
    logits,labels = eval_pred
    probs = torch.softmax(
        torch.tensor(logits),dim=-1)[:,1].numpy()
    preds = (probs>=0.5).astype(int)
    p,r,_ = precision_recall_curve(labels,probs)
    return {'pr_auc':round(auc(r,p),4),
            'f1':round(f1_score(labels,preds),4)}

args = TrainingArguments(
    output_dir='./bert_out',
    num_train_epochs=3,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=64,
    warmup_ratio=0.1, weight_decay=0.01,
    eval_strategy='epoch',
    save_strategy='epoch',
    load_best_model_at_end=True,
    metric_for_best_model='f1',
    logging_dir='./logs', report_to='none',
    fp16=torch.cuda.is_available()
)

trainer = Trainer(
    model=bert_model, args=args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    compute_metrics=compute_metrics
)

print("⏳ Training BERT (~15 mins GPU)...")
trainer.train()
bert_results = trainer.evaluate()
bert_pr  = bert_results.get('eval_pr_auc',0)
bert_f1  = bert_results.get('eval_f1',0)
print(f"\n📊 BERT Results:")
print(f"  PR-AUC: {bert_pr}")
print(f"  F1    : {bert_f1}")
trainer.save_model('./bert_best')
print("\n✅ CELL 7 done!")
```

```python
# ══════════════════════════════════════════════
# CELL 9: LIVE DETECTOR
# All 4 models vote together
# ══════════════════════════════════════════════

def predict_all_models(text):
    """
    Paper-based ensemble:
    CNN + LSTM + Hybrid + Attention + BERT
    """
    clean = clean_text(text)
    seq   = tokenizer_keras.texts_to_sequences(
        [clean])
    padded = pad_sequences(seq,
                            maxlen=MAX_LEN,
                            padding='post')

    # TF model predictions
    p_cnn  = float(cnn_model.predict(
        padded, verbose=0)[0][0])
    p_lstm = float(lstm_model.predict(
        padded, verbose=0)[0][0])
    p_hyb  = float(hybrid_model.predict(
        padded, verbose=0)[0][0])
    p_attn = float(attn_model.predict(
        padded, verbose=0)[0][0])

    # BERT prediction
    try:
        inp_bert = bert_tokenizer(
            clean, return_tensors='pt',
            padding='max_length',
            truncation=True, max_length=64)
        with torch.no_grad():
            out = bert_model(**inp_bert)
        p_bert = torch.softmax(
            out.logits,dim=1)[0][1].item()
    except:
        p_bert = (p_cnn+p_lstm+p_hyb+p_attn)/4

    # Ensemble average
    avg = (p_cnn+p_lstm+p_hyb+p_attn+p_bert)/5

    if   avg >= 0.75: v="🔴 FAKE NEWS"
    elif avg >= 0.55: v="🟠 शायद FAKE"
    elif avg >= 0.45: v="🟡 Uncertain"
    elif avg >= 0.25: v="🟡 शायद REAL"
    else:             v="🟢 REAL NEWS"

    print("\n"+"═"*55)
    print(f"📰 {text[:60]}")
    print("─"*55)
    print(f"  CNN Model     : {p_cnn:.1%}")
    print(f"  BiLSTM        : {p_lstm:.1%}")
    print(f"  Hybrid CNN+LSTM: {p_hyb:.1%}")
    print(f"  Attention     : {p_attn:.1%}")
    print(f"  BERT          : {p_bert:.1%}")
    print("─"*55)
    print(f"  Ensemble Avg  : {avg:.1%}")
    print(f"  🎯 VERDICT    : {v}")
    print("═"*55)
    return avg, v

# Test
print("LIVE DETECTOR TEST:")
tests = [
    "BREAKING: Celebrity secretly arrested SHOCKING!",
    "Scientists discover new cancer treatment",
    "RCB beat SRH in yesterday IPL match",
    "Messi scored 4 goals in today match",
    "Parliament passes new education policy",
]
for t in tests:
    predict_all_models(t)

# Interactive
print("\n" + "="*45)
print("Type any news to check!")
print("Type 'quit' to stop")
print("="*45)

while True:
    text = input("\n📝 News: ").strip()
    if text.lower() == 'quit': break
    if len(text) > 5:
        predict_all_models(text)
```

```python
# ══════════════════════════════════════════════
# GRADIO FAKE NEWS DETECTOR
# Image upload + Text input dono supported
# ══════════════════════════════════════════════

!pip install -q gradio pytesseract pillow
!apt-get install -q -y tesseract-ocr

import gradio as gr
import numpy as np, pickle, re, torch, io
from tensorflow.keras.preprocessing.sequence import pad_sequences
from PIL import Image, ImageFilter, ImageEnhance
import pytesseract

MAX_LEN = 50

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def predict_from_text(text):
    if not text or len(text.strip()) < 5:
        return "⚠️ Thoda lamba text likho!", "", "", "", "", "", ""

    clean = clean_text(text)
    seq   = tokenizer_keras.texts_to_sequences([clean])
    padded = pad_sequences(seq, maxlen=MAX_LEN, padding='post')

    p_cnn  = float(cnn_model.predict(padded,  verbose=0)[0][0])
    p_lstm = float(lstm_model.predict(padded,  verbose=0)[0][0])
    p_hyb  = float(hybrid_model.predict(padded,verbose=0)[0][0])
    p_attn = float(attn_model.predict(padded,  verbose=0)[0][0])

    try:
        inp_bert = bert_tokenizer(
            clean, return_tensors='pt',
            padding='max_length', truncation=True, max_length=64)
        with torch.no_grad():
            out = bert_model(**inp_bert)
        p_bert = torch.softmax(out.logits, dim=1)[0][1].item()
    except:
        p_bert = (p_cnn + p_lstm + p_hyb + p_attn) / 4

    avg = (p_cnn + p_lstm + p_hyb + p_attn + p_bert) / 5

    if   avg >= 0.75: verdict = "🔴 FAKE NEWS"
    elif avg >= 0.55: verdict = "🟠  FAKE"
    elif avg >= 0.45: verdict = "🟡 Uncertain"
    elif avg >= 0.25: verdict = "🟡  REAL"
    else:             verdict = "🟢 REAL NEWS"

    return (
        verdict,
        f"{p_cnn:.1%}",
        f"{p_lstm:.1%}",
        f"{p_hyb:.1%}",
        f"{p_attn:.1%}",
        f"{p_bert:.1%}",
        f"{avg:.1%}"
    )

def predict_from_image(image):
    if image is None:
        return "⚠️ Pehle image upload karo!", "", "", "", "", "", ""

    # PIL Image → OCR
    img = image.convert('L')
    img = img.filter(ImageFilter.SHARPEN)
    img = ImageEnhance.Contrast(img).enhance(2.0)

    text = pytesseract.image_to_string(img, config='--psm 6').strip()

    if len(text) < 10:
        return "❌ Image mein text nahi mila! Clear screenshot use karo.", "", "", "", "", "", ""

    # Reuse text predictor
    return predict_from_text(text)

# ── Gradio UI ──
with gr.Blocks(title="Fake News Detector - Made By Pralhad", theme=gr.themes.Soft()) as demo:

    gr.Markdown("""
    # 📰 Fake News Detector
    ### Deep Learning Ensemble — CNN + BiLSTM + Hybrid + Attention + BERT
    """)

    with gr.Tabs():

        # ── TAB 1: Text Input ──
        with gr.TabItem("📝 Text se Check karo"):
            with gr.Row():
                with gr.Column(scale=2):
                    text_input = gr.Textbox(
                        label="News headline ya text yahan likho",
                        placeholder="e.g. BREAKING: Celebrity secretly arrested SHOCKING!",
                        lines=4
                    )
                    text_btn = gr.Button("🔍 Check Karo", variant="primary")

                with gr.Column(scale=2):
                    text_verdict = gr.Textbox(label="🎯 Verdict", interactive=False)
                    with gr.Row():
                        t_cnn  = gr.Textbox(label="CNN",      interactive=False)
                        t_lstm = gr.Textbox(label="BiLSTM",   interactive=False)
                        t_hyb  = gr.Textbox(label="Hybrid",   interactive=False)
                    with gr.Row():
                        t_attn = gr.Textbox(label="Attention",interactive=False)
                        t_bert = gr.Textbox(label="BERT",     interactive=False)
                        t_avg  = gr.Textbox(label="Ensemble", interactive=False)

            text_btn.click(
                fn=predict_from_text,
                inputs=text_input,
                outputs=[text_verdict, t_cnn, t_lstm, t_hyb, t_attn, t_bert, t_avg]
            )

            gr.Examples(
                examples=[
                    ["SHOCKING: Celebrity secretly arrested nobody knew!"],
                    ["Scientists discover new cancer treatment in clinical trials"],
                    ["Parliament passes new education policy today"],
                    ["BREAKING: Messi SECRETLY signed with new club CONFIRMED sources say"],
                    ["India wins test series against Australia by 3 wickets"],
                ],
                inputs=text_input
            )

        # ── TAB 2: Image Upload ──
        with gr.TabItem("📸 Image se Check karo"):
            with gr.Row():
                with gr.Column(scale=2):
                    img_input = gr.Image(
                        label="Article ka screenshot ya photo upload karo",
                        type="pil"
                    )
                    img_btn = gr.Button("🔍 Image Analyze Karo", variant="primary")

                with gr.Column(scale=2):
                    img_verdict = gr.Textbox(label="🎯 Verdict", interactive=False)
                    with gr.Row():
                        i_cnn  = gr.Textbox(label="CNN",      interactive=False)
                        i_lstm = gr.Textbox(label="BiLSTM",   interactive=False)
                        i_hyb  = gr.Textbox(label="Hybrid",   interactive=False)
                    with gr.Row():
                        i_attn = gr.Textbox(label="Attention",interactive=False)
                        i_bert = gr.Textbox(label="BERT",     interactive=False)
                        i_avg  = gr.Textbox(label="Ensemble", interactive=False)

            img_btn.click(
                fn=predict_from_image,
                inputs=img_input,
                outputs=[img_verdict, i_cnn, i_lstm, i_hyb, i_attn, i_bert, i_avg]
            )

            gr.Markdown("""
            **Tips for best results:**
            - Clear screenshot use karo
            - News headline visible honi chahiye
            - Blurry ya dark images avoid karo
            """)

    gr.Markdown("""
    ---
    **Model Info:** CNN (Paper [17]) | BiLSTM (Paper [4]) | Hybrid CNN+LSTM (Paper [19]) | Attention+BERT (Paper [31,14])
    **Dataset:** FakeNewsNet — GossipCop + PolitiFact | **Paper:** Al-Tai et al. (2023)
    """)

demo.launch(share=True, debug=False)
```

```python
# ══════════════════════════════════════════════
# CELL 9: GRADIO MULTIMODAL FAKE NEWS DETECTOR
# Upload Image + Type Text → Check Fake/Real
# Web Search + CLIP + BERT + OCR
# ══════════════════════════════════════════════

!pip install -q gradio
!pip install -q duckduckgo-search
!pip install -q pytesseract pillow
!apt-get install -q tesseract-ocr
!pip install -q transformers torch

import gradio as gr
import torch
import numpy as np
import requests
from PIL import Image
import pytesseract
import io
import re
import warnings
import time
warnings.filterwarnings('ignore')

from duckduckgo_search import DDGS
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    CLIPProcessor,
    CLIPModel
)
from scipy.sparse import hstack, csr_matrix

# ──────────────────────────────────────────────
# LOAD ALL MODELS
# ──────────────────────────────────────────────
print("⏳ Loading models...")

# BERT
bert_tok = AutoTokenizer.from_pretrained(
    'distilbert-base-uncased')
try:
    bert_mdl = AutoModelForSequenceClassification\
        .from_pretrained('./bert_best')
    print("✓ Fine-tuned BERT!")
except:
    bert_mdl = AutoModelForSequenceClassification\
        .from_pretrained(
            'distilbert-base-uncased',
            num_labels=2)
    print("⚠️  Base BERT")
bert_mdl.eval()

# CLIP
print("⏳ Loading CLIP...")
clip_mdl  = CLIPModel.from_pretrained(
    "openai/clip-vit-base-patch32")
clip_proc = CLIPProcessor.from_pretrained(
    "openai/clip-vit-base-patch32")
clip_mdl.eval()
print("✓ All models ready!\n")

URGENCY = ['breaking','urgent','alert','leaked',
           'confirmed','exposed','shocking','must',
           'secret','exclusive','banned','censored',
           'truth','bombshell','revealed','hoax',
           'fake','fraud','scam']

TRUSTED = ['bbc.com','reuters.com','apnews.com',
           'theguardian.com','nytimes.com',
           'wikipedia.org','britannica.com',
           'espn.com','cricbuzz.com','ndtv.com',
           'thehindu.com','indiatoday.in',
           'nationalgeographic.com','nasa.gov',
           'who.int','un.org','iplt20.com',
           'fifa.com','uefa.com','bcci.tv']

# ──────────────────────────────────────────────
# HELPER FUNCTIONS
# ──────────────────────────────────────────────

def get_bert_score(text):
    inp = bert_tok(text,
                   return_tensors='pt',
                   padding='max_length',
                   truncation=True,
                   max_length=64)
    with torch.no_grad():
        out = bert_mdl(**inp)
    return torch.softmax(
        out.logits,dim=1)[0][1].item()


def get_clip_score(image, text):
    try:
        descs = [
            text,
            f"a photo of {text}",
            "a real authentic photo",
            "a fake misleading photo",
            "a manipulated image",
        ]
        inp = clip_proc(
            text=descs,
            images=image,
            return_tensors="pt",
            padding=True)
        with torch.no_grad():
            out   = clip_mdl(**inp)
            probs = out.logits_per_image\
                       .softmax(dim=1)[0]
        return {d:float(p)
                for d,p in zip(descs,probs)}
    except:
        return {}


def do_ocr(image):
    try:
        text = pytesseract.image_to_string(
            image, config='--psm 6').strip()
        return text if text else None
    except:
        return None


def web_search(query):
    trusted_found = []
    all_found     = []
    snippets      = []
    try:
        with DDGS() as ddgs:
            results = list(ddgs.text(
                query, max_results=6))
        for r in results:
            url     = r.get('href','')
            title   = r.get('title','')
            body    = r.get('body','')[:120]
            all_found.append(
                {'url':url,'title':title,
                 'body':body})
            snippets.append(body)
            if any(s in url for s in TRUSTED):
                trusted_found.append(
                    {'url':url,'title':title,
                     'body':body})
    except:
        pass
    return trusted_found, all_found, snippets


def get_text_flags(text):
    t     = text.lower()
    words = t.split()
    flags = []
    if sum(1 for w in words
           if w in URGENCY) > 0:
        flags.append("⚠️ Urgency/clickbait words")
    if '!' in t:
        flags.append("⚠️ Exclamation mark")
    if sum(1 for c in text
           if c.isupper())/max(1,len(text))>0.15:
        flags.append("⚠️ Many CAPS letters")
    if '?' in t:
        flags.append("⚠️ Question mark bait")
    if len(words) > 20:
        flags.append(f"⚠️ Very long title "
                     f"({len(words)} words)")
    return flags


# ──────────────────────────────────────────────
# MAIN DETECTION FUNCTION
# ──────────────────────────────────────────────

def fake_news_detector(claim_text, image):
    """
    Main function called by Gradio
    Returns: verdict, score bar,
             detailed report
    """

    if not claim_text or len(claim_text.strip())<5:
        return (
            "⚠️ Please enter a claim/news text!",
            0,
            "No text provided."
        )

    claim_text = claim_text.strip()
    report     = []
    scores     = {}
    findings   = []

    report.append(
        f"# 🔍 Fake News Analysis Report\n")
    report.append(
        f"**Claim:** {claim_text}\n")
    report.append("---")

    # ──────────────────────────
    # STEP 1: BERT Text Analysis
    # ──────────────────────────
    report.append("## 🤖 Step 1: Text Analysis (BERT)")
    bert_score = get_bert_score(claim_text)
    scores['bert'] = bert_score
    text_flags     = get_text_flags(claim_text)

    report.append(
        f"- BERT Fake Score: **{bert_score:.1%}**")
    if text_flags:
        report.append("- Red Flags Found:")
        for f in text_flags:
            report.append(f"  - {f}")
            findings.append(f)
    else:
        report.append(
            "- ✅ No text red flags found")
    report.append("")

    # ──────────────────────────
    # STEP 2: Image Analysis
    # ──────────────────────────
    image_mismatch = 0.0

    if image is not None:
        report.append(
            "## 🖼️ Step 2: Image Analysis")
        pil_img = Image.fromarray(image)\
            if not isinstance(image,Image.Image)\
            else image

        # OCR
        ocr_text = do_ocr(pil_img)
        if ocr_text:
            report.append(
                f"- 📝 OCR Text found in image:"
                f" `{ocr_text[:80]}`")
            findings.append(
                f"Image contains text: "
                f"'{ocr_text[:50]}'")

            # OCR text vs claim match
            claim_words = set(
                claim_text.lower().split())
            ocr_words   = set(
                ocr_text.lower().split())
            overlap     = claim_words & ocr_words
            if len(overlap) < 2:
                findings.append(
                    "⚠️ OCR text doesn't match claim!")
        else:
            report.append(
                "- 📝 OCR: No text found in image")

        # CLIP
        report.append(
            "\n## 🔗 Step 3: "
            "Image-Text Matching (CLIP)")
        clip_scores = get_clip_score(
            pil_img, claim_text)

        if clip_scores:
            claim_match = clip_scores.get(
                claim_text, 0.5)
            scores['clip'] = claim_match

            report.append(
                f"- Claim match score: "
                f"**{claim_match:.1%}**")
            report.append(
                "- Top matches:")
            sorted_s = sorted(
                clip_scores.items(),
                key=lambda x: x[1],
                reverse=True)[:3]
            for desc,prob in sorted_s:
                bar = "█" * int(prob*20)
                report.append(
                    f"  - `{desc[:40]}`: "
                    f"{prob:.1%} {bar}")

            # Low match = mismatch
            if claim_match < 0.20:
                image_mismatch = 0.90
                msg = ("⚠️ **Image does NOT match"
                       " the claim!** "
                       "Possible FAKE!")
                report.append(f"\n- {msg}")
                findings.append(
                    "Image-claim MISMATCH detected!")
            elif claim_match < 0.35:
                image_mismatch = 0.60
                findings.append(
                    "Low image-claim similarity")
            else:
                image_mismatch = 0.10
                report.append(
                    "- ✅ Image matches claim")

        scores['image_mismatch'] = image_mismatch
        report.append("")

    else:
        report.append(
            "## 🖼️ Step 2 & 3: Image\n"
            "- ℹ️ No image provided — "
            "text + web analysis only\n")

    # ──────────────────────────
    # STEP 4: Web Search
    # ──────────────────────────
    report.append(
        "## 🌐 Step 4: Web Fact-Check")
    trusted, all_res, snippets = web_search(
        claim_text)
    n_trusted = len(trusted)
    n_total   = len(all_res)

    if   n_trusted >= 3: web_score = 0.05
    elif n_trusted == 2: web_score = 0.15
    elif n_trusted == 1: web_score = 0.30
    elif n_total   >= 4: web_score = 0.50
    elif n_total   >= 1: web_score = 0.70
    else:                web_score = 0.90

    scores['web'] = web_score
    report.append(
        f"- Results: **{n_total} total**, "
        f"**{n_trusted} trusted sources**")

    if trusted:
        report.append(
            "\n✅ **Trusted Sources Found:**")
        for s in trusted[:3]:
            report.append(
                f"  - [{s['title'][:45]}]"
                f"({s['url']})")
            if s.get('body'):
                report.append(
                    f"    > {s['body'][:100]}...")
    elif all_res:
        report.append(
            "\n⚠️ No trusted sources. "
            "Other results:")
        for r in all_res[:3]:
            report.append(
                f"  - {r['title'][:50]}")
            if r.get('body'):
                report.append(
                    f"    > {r['body'][:80]}...")
    else:
        report.append(
            "\n❌ **No results found!** "
            "Very suspicious!")
        findings.append(
            "No web sources found for this claim")

    report.append("")

    # ──────────────────────────
    # STEP 5: Combined Score
    # ──────────────────────────
    report.append(
        "## 📊 Step 5: Combined Analysis")

    if image is not None:
        final = (
            scores['bert']           * 0.20 +
            scores['image_mismatch'] * 0.45 +
            scores['web']            * 0.35
        )
        method = ("BERT(20%) + "
                  "Image Mismatch(45%) + "
                  "Web(35%)")
    else:
        final = (
            scores['bert'] * 0.35 +
            scores['web']  * 0.65
        )
        method = "BERT(35%) + Web(65%)"

    report.append(f"- Method: `{method}`")
    report.append(
        f"- BERT Score: {scores['bert']:.1%}")
    if 'image_mismatch' in scores:
        report.append(
            f"- Image Mismatch: "
            f"{scores['image_mismatch']:.1%}")
    report.append(
        f"- Web Score: {scores['web']:.1%}")
    report.append(
        f"- **Final Score: {final:.1%}**")
    report.append("")

    # ──────────────────────────
    # VERDICT
    # ──────────────────────────
    if   final >= 0.75:
        verdict = "🔴 FAKE NEWS"
        color   = "red"
        conf    = "Very high chance this is FAKE"
    elif final >= 0.60:
        verdict = "🟠 Likely FAKE"
        color   = "orange"
        conf    = "Suspicious — please verify"
    elif final >= 0.45:
        verdict = "🟡 Uncertain"
        color   = "yellow"
        conf    = "Cannot confirm — check manually"
    elif final >= 0.25:
        verdict = "🟡 Likely REAL"
        color   = "lightgreen"
        conf    = "Probably real news"
    else:
        verdict = "🟢 REAL NEWS"
        color   = "green"
        conf    = "Sources verified — appears real"

    report.append("---")
    report.append(f"# {verdict}")
    report.append(f"**{conf}**")
    report.append(
        f"\n**Confidence Score: {final:.1%}**")

    if findings:
        report.append("\n## 🔍 Key Findings:")
        for f in findings:
            report.append(f"- {f}")

    report.append("\n---")
    report.append(
        "*Analysis by: BERT + CLIP + OCR + "
        "DuckDuckGo Web Search*")

    score_pct = int(final * 100)

    return (
        verdict,
        score_pct,
        "\n".join(report)
    )


# ──────────────────────────────────────────────
# GRADIO INTERFACE
# ──────────────────────────────────────────────

with gr.Blocks(
    title="Fake News Detector",
    theme=gr.themes.Soft()
) as demo:

    gr.Markdown("""
    # 🔍 Multimodal Fake News Detector
    ### Text + Image + Web Search = Complete Detection
    > **How to use:**
    > 1. Type any news claim or social media post
    > 2. Optionally upload an image
    > 3. Click **Analyze** to check if it's fake or real
    """)

    with gr.Row():
        # Left column — inputs
        with gr.Column(scale=1):
            gr.Markdown("### 📝 Input")

            claim_input = gr.Textbox(
                label="News Claim / Social Media Post",
                placeholder=(
                    "e.g. Messi scored 4 goals "
                    "in today's match\n"
                    "e.g. This is Mount Everest\n"
                    "e.g. RCB beat SRH yesterday"),
                lines=4
            )

            image_input = gr.Image(
                label="Upload Image (Optional)",
                type="numpy",
                height=250
            )

            analyze_btn = gr.Button(
                "🔍 Analyze",
                variant="primary",
                size="lg"
            )

            gr.Markdown("### 📌 Quick Examples")
            gr.Examples(
                examples=[
                    ["Messi scored 4 goals in today's match",
                     None],
                    ["BREAKING: Celebrity secretly "
                     "arrested SHOCKING truth!",
                     None],
                    ["Scientists discover new "
                     "cancer treatment",None],
                    ["RCB beat SRH in yesterday's "
                     "IPL match",None],
                    ["This is Mount Everest",None],
                    ["LEAKED: Government hiding "
                     "alien contact BOMBSHELL!",None],
                ],
                inputs=[claim_input, image_input]
            )

        # Right column — outputs
        with gr.Column(scale=1):
            gr.Markdown("### 📊 Results")

            verdict_out = gr.Textbox(
                label="🎯 Verdict",
                interactive=False,
                text_align="center"
            )

            score_out = gr.Slider(
                label="Fake News Probability (%)",
                minimum=0,
                maximum=100,
                value=0,
                interactive=False,
                info="0% = Definitely Real | "
                     "100% = Definitely Fake"
            )

            report_out = gr.Markdown(
                label="📋 Detailed Report",
                value="*Analysis will appear here...*"
            )

    # Connect button
    analyze_btn.click(
        fn=fake_news_detector,
        inputs=[claim_input, image_input],
        outputs=[verdict_out,
                 score_out,
                 report_out]
    )

    gr.Markdown("""
    ---
    ### 📖 How It Works
    | Step | Method | Purpose |
    |------|--------|---------|
    | 1 | BERT (Deep Learning) | Text pattern analysis |
    | 2 | OCR (Tesseract) | Extract text from image |
    | 3 | CLIP (OpenAI) | Image-text matching |
    | 4 | DuckDuckGo Search | Real-time fact checking |
    | 5 | Weighted Ensemble | Combined final decision |

    ### 🎯 Score Guide
    - 🟢 **0-25%** = Real News
    - 🟡 **25-45%** = Probably Real
    - 🟡 **45-60%** = Uncertain
    - 🟠 **60-75%** = Likely Fake
    - 🔴 **75-100%** = Fake News

    *Dataset: FakeNewsNet (GossipCop + PolitiFact)*
    *Paper: Al-Tai et al. 2023 — Deep Learning for Fake News Detection*
    """)

# Launch
print("\n🚀 Launching Gradio app...")
demo.launch(
    share=True,      # Public URL मिलेगा
    debug=False,
    quiet=True
)
```

```python
# ═══════════════════════════════════════════════════════════════════════════
# SMART FAKE NEWS DETECTOR — FULLY CORRECTED VERSION
# अब फेक इमेज को REAL नहीं बताएगा
# ═══════════════════════════════════════════════════════════════════════════

!pip install -q gradio duckduckgo-search pillow pytesseract transformers torch newspaper3k requests beautifulsoup4 lxml_html_clean easyocr

import gradio as gr
import torch
import numpy as np
import requests
from PIL import Image, ImageFilter, ImageEnhance, ImageStat, ImageChops
import pytesseract
import io
import re
import warnings
import time
import json
from bs4 import BeautifulSoup
warnings.filterwarnings('ignore')

# EasyOCR (better than Tesseract)
try:
    import easyocr
    EASYOCR_AVAILABLE = True
    easyocr_reader = easyocr.Reader(['en'])
    print("✓ EasyOCR loaded (better text extraction)")
except:
    EASYOCR_AVAILABLE = False
    print("⚠️ EasyOCR not available, using Tesseract")

try:
    from duckduckgo_search import DDGS
    DDGS_AVAILABLE = True
except:
    DDGS_AVAILABLE = False

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    CLIPProcessor,
    CLIPModel,
    pipeline
)

# ──────────────────────────────────────────────
# MODELS LOAD KARO
# ──────────────────────────────────────────────
print("⏳ Sab models load ho rahe hain...")

# BERT for text fake detection
bert_tok = AutoTokenizer.from_pretrained('distilbert-base-uncased')
try:
    bert_mdl = AutoModelForSequenceClassification.from_pretrained('./bert_best')
    print("✓ Fine-tuned BERT loaded!")
except:
    bert_mdl = AutoModelForSequenceClassification.from_pretrained(
        'distilbert-base-uncased', num_labels=2)
    print("⚠️ Base BERT (fine-tuned nahi mila)")
bert_mdl.eval()

# CLIP for image understanding
print("⏳ CLIP load ho raha hai...")
clip_mdl  = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_proc = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
clip_mdl.eval()

# Image captioning
print("⏳ Image captioning model load ho raha hai...")
try:
    captioner = pipeline(
        "image-to-text",
        model="Salesforce/blip-image-captioning-base",
        max_new_tokens=50
    )
    CAPTION_AVAILABLE = True
    print("✓ Image captioner ready!")
except Exception as e:
    print(f"⚠️ Captioner nahi chala: {e}")
    CAPTION_AVAILABLE = False

print("✓ Sab models ready!\n")

# ──────────────────────────────────────────────
# CONSTANTS
# ──────────────────────────────────────────────
TRUSTED_SITES = [
    'bbc.com', 'reuters.com', 'apnews.com',
    'theguardian.com', 'nytimes.com',
    'wikipedia.org', 'britannica.com',
    'espn.com', 'cricbuzz.com', 'ndtv.com',
    'thehindu.com', 'indiatoday.in',
    'nationalgeographic.com', 'nasa.gov',
    'who.int', 'un.org', 'iplt20.com',
    'fifa.com', 'uefa.com', 'bcci.tv',
    'snopes.com', 'factcheck.org',
    'politifact.com', 'altnews.in',
    'boomlive.in', 'vishvasnews.com'
]

FAKE_WORDS = [
    'breaking', 'urgent', 'alert', 'leaked',
    'confirmed', 'exposed', 'shocking', 'must',
    'secret', 'exclusive', 'banned', 'censored',
    'bombshell', 'revealed', 'hoax', 'fraud',
    'scam', 'conspiracy', 'hidden truth',
    'they dont want you to know', 'wake up',
    'share before deleted', 'going viral'
]

# ──────────────────────────────────────────────
# HELPER FUNCTIONS
# ──────────────────────────────────────────────

def clean_text(text):
    """Text clean karo"""
    text = str(text).lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def get_bert_score(text):
    """BERT se fake score lo"""
    if not text or len(text.strip()) < 3:
        return 0.5
    inp = bert_tok(
        text, return_tensors='pt',
        padding='max_length', truncation=True, max_length=64
    )
    with torch.no_grad():
        out = bert_mdl(**inp)
    return torch.softmax(out.logits, dim=1)[0][1].item()

def get_image_caption(pil_img):
    """AI se image ka description banao"""
    if not CAPTION_AVAILABLE:
        return None
    try:
        result = captioner(pil_img)
        if result and len(result) > 0:
            return result[0].get('generated_text', '').strip()
    except Exception as e:
        print(f"Caption error: {e}")
    return None

def do_ocr(pil_img):
    """Image se text nikalo - EasyOCR preferred, fallback Tesseract"""
    try:
        # EasyOCR (better)
        if EASYOCR_AVAILABLE:
            img_np = np.array(pil_img)
            result = easyocr_reader.readtext(img_np, detail=0, paragraph=True)
            if result:
                text = ' '.join(result)
                text = re.sub(r'\s+', ' ', text).strip()
                if len(text) > 5:
                    return text

        # Fallback: Tesseract
        img_gray = pil_img.convert('L')
        img_sharp = img_gray.filter(ImageFilter.SHARPEN)
        img_enhance = ImageEnhance.Contrast(img_sharp).enhance(2.0)
        texts = []
        for psm in [6, 3, 11]:
            try:
                t = pytesseract.image_to_string(img_enhance, config=f'--psm {psm}').strip()
                if t and len(t) > 5:
                    texts.append(t)
            except:
                pass
        if texts:
            best = max(texts, key=len)
            best = re.sub(r'\n+', ' ', best)
            best = re.sub(r'\s+', ' ', best).strip()
            return best if len(best) > 3 else None
    except Exception as e:
        print(f"OCR error: {e}")
    return None

def get_clip_understanding(pil_img):
    """CLIP se image ka context samjho - enhanced with manipulation categories"""
    try:
        categories = [
            # News categories
            "a news photograph",
            "a social media screenshot",
            "a meme or fake image",
            "a real authentic photograph",
            "a digitally manipulated image",
            "a political image",
            "a sports photograph",
            "a disaster or crisis photograph",
            "a celebrity photograph",
            "a protest or rally photograph",
            # Locations
            "a mountain landscape",
            "a city building",
            "a natural disaster",
            "a monument or landmark",
            # NEW: specific manipulation categories
            "a real photograph with no text added",
            "a photograph with fake text added on top",
            "a doctored or manipulated news image",
            "a meme with false claim",
            "an AI generated image",
            "a photoshopped image",
        ]
        inputs = clip_proc(
            text=categories,
            images=pil_img,
            return_tensors="pt",
            padding=True
        )
        with torch.no_grad():
            outputs = clip_mdl(**inputs)
            probs = outputs.logits_per_image.softmax(dim=1)[0]
        results = {cat: float(prob) for cat, prob in zip(categories, probs)}
        top3 = sorted(results.items(), key=lambda x: x[1], reverse=True)[:3]
        return results, top3
    except Exception as e:
        print(f"CLIP error: {e}")
        return {}, []

def check_image_manipulation(pil_img):
    """Image manipulation signs dhundo - enhanced with ELA"""
    findings = []
    score = 0.0
    try:
        # ELA (Error Level Analysis)
        def ela_score(img, quality=90):
            try:
                from io import BytesIO
                img_rgb = img.convert('RGB')
                output = BytesIO()
                img_rgb.save(output, format='JPEG', quality=quality)
                output.seek(0)
                recompressed = Image.open(output)
                diff = ImageChops.difference(img_rgb, recompressed)
                extrema = diff.getextrema()
                max_diff = max([max(pair) for pair in extrema]) if extrema else 0
                return max_diff
            except:
                return 0
        ela_val = ela_score(pil_img)
        if ela_val > 30:   # High difference -> possible manipulation
            findings.append(f"⚠️ ELA high ({ela_val:.0f}) – compression inconsistency")
            score += 0.3

        # Image statistics
        img_rgb = pil_img.convert('RGB')
        stat = ImageStat.Stat(img_rgb)
        r, g, b = stat.stddev[:3]
        if max(r, g, b) < 15:
            findings.append("⚠️ Very uniform colors — possibly AI generated")
            score += 0.3
        # Noise / smoothness
        img_array = np.array(img_rgb).astype(float)
        h, w = img_array.shape[:2]
        patch_size = min(50, h//4, w//4)
        if patch_size > 10:
            patches_std = []
            for i in range(0, h - patch_size, patch_size):
                for j in range(0, w - patch_size, patch_size):
                    patch = img_array[i:i+patch_size, j:j+patch_size]
                    patches_std.append(np.std(patch))
            avg_std = np.mean(patches_std)
            if avg_std < 8:
                findings.append("⚠️ Unusually smooth texture — possible AI/CGI image")
                score += 0.25
        # Resolution check
        width, height = pil_img.size
        if width < 200 or height < 200:
            findings.append("⚠️ Very low resolution image")
            score += 0.1
        # Aspect ratio
        ratio = width / height
        if ratio > 4 or ratio < 0.25:
            findings.append("⚠️ Unusual aspect ratio")
            score += 0.1
    except Exception as e:
        print(f"Manipulation check error: {e}")
    return findings, min(score, 1.0)

def web_search_and_read(query, max_results=8):
    """
    Internet search + support/contradict analysis.
    Returns: trusted_sources, all_sources, supporting_count, contradicting_count
    """
    trusted_sources = []
    all_sources = []
    supporting = 0
    contradicting = 0

    if not DDGS_AVAILABLE:
        return [], [], 0, 0

    try:
        with DDGS() as ddgs:
            results = list(ddgs.text(query, max_results=max_results))

        # Extract claim keywords from query (exclude common noise)
        claim_keywords = [w for w in query.lower().split()
                         if len(w) > 4 and w not in ['hoax', 'fake', 'fact', 'check', 'false', 'myth']]

        for r in results:
            url = r.get('href', '')
            title = r.get('title', '')
            body = r.get('body', '')[:300]
            is_trusted = any(s in url for s in TRUSTED_SITES)

            # Combine text for analysis
            full_text = (title + " " + body).lower()

            # Check for contradiction (source says it's false)
            contradict_keywords = ['false', 'hoax', 'fake', 'debunked', 'never happened',
                                   'not true', 'no evidence', 'myth', 'still standing',
                                   'not demolished', 'still exists']
            contradicts = any(kw in full_text for kw in contradict_keywords)

            # Check for support (source confirms the claim)
            supports = any(kw in full_text for kw in claim_keywords) and not contradicts

            if supports and is_trusted:
                supporting += 1
            if contradicts and is_trusted:
                contradicting += 1

            all_sources.append({'url': url, 'title': title, 'body': body, 'trusted': is_trusted})
            if is_trusted:
                trusted_sources.append({'url': url, 'title': title, 'body': body})

        print(f"DEBUG: supporting={supporting}, contradicting={contradicting}")

    except Exception as e:
        print(f"Search error: {e}")

    return trusted_sources, all_sources, supporting, contradicting

def check_claim_vs_image(image_caption, ocr_text, clip_top3):
    """Image aur text mein match check karo"""
    issues = []
    mismatch = 0.0
    if not image_caption:
        return issues, 0.3
    caption_lower = image_caption.lower()
    if clip_top3:
        top_label = clip_top3[0][0].lower()
        if any(w in top_label for w in ['manipulated', 'photoshop', 'ai generated', 'doctored', 'fake text']):
            issues.append(f"🚨 AI/CLIP detects: '{clip_top3[0][0]}' (score: {clip_top3[0][1]:.1%})")
            mismatch += 0.5
        elif 'fake' in top_label or 'meme' in top_label:
            issues.append(f"⚠️ Image looks like: '{clip_top3[0][0]}'")
            mismatch += 0.3
    if ocr_text:
        ocr_lower = ocr_text.lower()
        fake_found = [w for w in FAKE_WORDS if w in ocr_lower]
        if fake_found:
            issues.append(f"⚠️ Suspicious words in image text: {fake_found}")
            mismatch += 0.2 * len(fake_found)
    return issues, min(mismatch, 1.0)

# ══════════════════════════════════════════════
# MAIN DETECTION FUNCTION (CORRECTED)
# ══════════════════════════════════════════════

def analyze_image_only(image):
    if image is None:
        return ("⚠️ Koi image upload nahi ki!", 0, "Pehle ek image upload karo.")

    report = []
    scores = {}
    all_findings = []
    start_time = time.time()

    if not isinstance(image, Image.Image):
        pil_img = Image.fromarray(image)
    else:
        pil_img = image

    report.append("# 🔍 Smart Fake News Analysis\n")
    report.append(f"Image size: {pil_img.size[0]}×{pil_img.size[1]} pixels\n")
    report.append("---\n")

    # Step 1: AI Caption
    report.append("## 👁️ Step 1: Image ko Samjho (AI Caption)")
    image_caption = get_image_caption(pil_img)
    if image_caption:
        report.append(f"- 🤖 AI Caption: **\"{image_caption}\"**")
        all_findings.append(f"Image shows: {image_caption}")
        search_query = image_caption
    else:
        report.append("- ⚠️ Caption generate nahi hua")
        search_query = "news fact check"
    report.append("")

    # Step 2: OCR
    report.append("## 📝 Step 2: Image Mein Text Dhundho (OCR)")
    ocr_text = do_ocr(pil_img)
    bert_score = 0.4
    if ocr_text and len(ocr_text) > 5:
        report.append(f"- ✓ Image mein text mila: **\"{ocr_text[:150]}\"**")
        all_findings.append(f"OCR text: {ocr_text[:100]}")
        # Build better search query: include OCR + "hoax"
        ocr_clean = re.sub(r'[^a-zA-Z0-9\s]', ' ', ocr_text)
        ocr_clean = re.sub(r'\s+', ' ', ocr_clean).strip()
        if len(ocr_clean) > 10:
            search_query = ocr_clean[:100] + " hoax OR fact check"
        bert_score = get_bert_score(clean_text(ocr_text[:200]))
        scores['bert_text'] = bert_score
        report.append(f"- 🤖 BERT Text Score: **{bert_score:.1%}** fake probability")
        fake_found = [w for w in FAKE_WORDS if w in ocr_text.lower()]
        if fake_found:
            report.append(f"- 🚨 Suspicious words mile: **{fake_found}**")
            all_findings.append(f"Suspicious words: {fake_found}")
        else:
            report.append("- ✅ Koi suspicious words nahi")
    else:
        report.append("- ℹ️ Image mein koi readable text nahi mila")
        scores['bert_text'] = 0.4
    report.append("")

    # Step 3: CLIP
    report.append("## 🔗 Step 3: Image Type Analysis (CLIP AI)")
    clip_all, clip_top3 = get_clip_understanding(pil_img)
    manipulation_score = 0.0
    if clip_top3:
        report.append("- Top matches:")
        for label, prob in clip_top3:
            bar = "█" * int(prob * 30)
            report.append(f"  - `{label}`: **{prob:.1%}** {bar}")
        top_label = clip_top3[0][0].lower()
        top_prob = clip_top3[0][1]
        if any(w in top_label for w in ['manipulated', 'photoshop', 'ai generated', 'fake', 'meme', 'doctored', 'fake text']):
            manipulation_score = top_prob * 0.8
            report.append(f"\n- 🚨 **WARNING: Image suspicious lagti hai!**")
            all_findings.append(f"CLIP says: '{clip_top3[0][0]}'")
        else:
            report.append(f"\n- ✅ Image type: '{clip_top3[0][0]}'")
        scores['clip_manipulation'] = manipulation_score
    report.append("")

    # Step 4: Manipulation check
    report.append("## 🔬 Step 4: Image Manipulation Check")
    manip_findings, manip_score = check_image_manipulation(pil_img)
    if manip_findings:
        for f in manip_findings:
            report.append(f"- {f}")
            all_findings.append(f)
        scores['manipulation'] = manip_score
        report.append(f"- Manipulation Score: **{manip_score:.1%}**")
    else:
        report.append("- ✅ No obvious manipulation detected")
        scores['manipulation'] = 0.1
    report.append("")

    # Step 5: Web Search with support/contradict
    report.append("## 🌐 Step 5: Internet Fact Check (Real-time)")
    report.append(f"- Searching for: **\"{search_query[:80]}\"**\n")
    trusted_src, all_src, supporting, contradicting = web_search_and_read(search_query)

    # NEW WEB SCORE LOGIC
    if contradicting >= 2:
        web_score = 0.95   # Strong evidence of fake
    elif supporting >= 2:
        web_score = 0.10   # Supports claim → likely real
    elif len(trusted_src) >= 3:
        web_score = 0.30
    elif len(all_src) >= 1:
        web_score = 0.70
    else:
        web_score = 0.90
    scores['web'] = web_score

    report.append(f"- 🔎 Total results: **{len(all_src)}**")
    report.append(f"- ✅ Trusted sources: **{len(trusted_src)}**")
    report.append(f"- 🏆 Sources supporting claim: **{supporting}**")
    report.append(f"- 🚨 Sources contradicting claim: **{contradicting}**")

    if trusted_src:
        report.append("\n### ✅ Trusted Sources:")
        for s in trusted_src[:4]:
            report.append(f"- 🔗 [{s['title'][:60]}]({s['url']})")
            if s.get('body'):
                report.append(f"  > {s['body'][:120]}...")
    if not all_src:
        report.append("\n- ❌ **Koi internet result nahi mila — bahut suspicious!**")
        all_findings.append("No internet results found")
    report.append("")

    # Step 6: Caption vs Internet match (semantic simplified)
    if image_caption and all_src:
        report.append("## 🔄 Step 6: Image vs Internet Match")
        caption_words = set(clean_text(image_caption).split())
        match_count = 0
        for src in all_src[:5]:
            src_text = clean_text(src.get('title', '') + ' ' + src.get('body', ''))
            src_words = set(src_text.split())
            overlap = caption_words & src_words
            if len(overlap) >= 2:
                match_count += 1
        match_ratio = match_count / min(5, len(all_src)) if all_src else 0
        scores['caption_web_match'] = 1.0 - match_ratio
        if match_ratio > 0.6:
            report.append(f"- ✅ Image content matches internet ({match_count}/{min(5, len(all_src))} sources agree)")
        elif match_ratio > 0.3:
            report.append(f"- ⚠️ Partial match ({match_count}/{min(5, len(all_src))} sources)")
            all_findings.append("Partial image-internet match")
        else:
            report.append(f"- 🚨 **Low match! Image content aur internet news mein fark!**")
            all_findings.append("Image content doesn't match internet news")
    else:
        scores['caption_web_match'] = 0.4
    report.append("")

    # Step 7: Final Score Calculation (adjusted weights)
    report.append("## 📊 Final Score Calculation")
    w_bert    = 0.15
    w_clip    = 0.15
    w_manip   = 0.15
    w_web     = 0.40      # increased from 0.35
    w_caption = 0.15      # decreased from 0.20

    final = (scores.get('bert_text', 0.4) * w_bert +
             scores.get('clip_manipulation', 0.1) * w_clip +
             scores.get('manipulation', 0.1) * w_manip +
             scores.get('web', 0.5) * w_web +
             scores.get('caption_web_match', 0.4) * w_caption)

    # If contradicting sources exist, boost final score (make it more fake)
    if contradicting >= 1:
        final = max(final, 0.65)
    if contradicting >= 2:
        final = max(final, 0.85)

    report.append(f"| Component | Score | Weight |")
    report.append(f"|-----------|-------|--------|")
    report.append(f"| BERT text | {scores.get('bert_text', 0.4):.1%} | 15% |")
    report.append(f"| CLIP manipulation | {scores.get('clip_manipulation', 0.1):.1%} | 15% |")
    report.append(f"| Image analysis | {scores.get('manipulation', 0.1):.1%} | 15% |")
    report.append(f"| Web fact-check | {scores.get('web', 0.5):.1%} | 40% |")
    report.append(f"| Caption match | {scores.get('caption_web_match', 0.4):.1%} | 15% |")
    report.append(f"| **FINAL** | **{final:.1%}** | 100% |")
    report.append("")

    # Verdict
    elapsed = time.time() - start_time
    if final >= 0.75:
        verdict = "🔴 FAKE NEWS"
        conf = "Bahut zyada chance hai ye FAKE hai!"
    elif final >= 0.60:
        verdict = "🟠 Likely FAKE"
        conf = "Suspicious — zaroor verify karo"
    elif final >= 0.45:
        verdict = "🟡 Uncertain"
        conf = "Confirm nahi ho sakta — manually check karo"
    elif final >= 0.25:
        verdict = "🟡 Likely REAL"
        conf = "Probably real news hai"
    else:
        verdict = "🟢 REAL NEWS"
        conf = "Sources verified — real lagta hai"

    report.append("---")
    report.append(f"# {verdict}")
    report.append(f"**{conf}**")
    report.append(f"\n**Final Score: {final:.1%}** | Analysis time: {elapsed:.1f}s")

    if all_findings:
        report.append("\n## 🔍 Key Findings:")
        for f in all_findings[:8]:
            report.append(f"- {f}")

    report.append("\n---")
    report.append("*Analysis: AI Caption + OCR (EasyOCR) + CLIP + Manipulation (ELA) + Web Search with Contradiction Check*")

    return verdict, int(final * 100), "\n".join(report)

# ══════════════════════════════════════════════
# GRADIO UI
# ══════════════════════════════════════════════

with gr.Blocks(title="Smart Fake News Detector - Corrected", theme=gr.themes.Soft()) as demo:
    gr.Markdown("""
    # 🔍 Smart Fake News Detector (CORRECTED VERSION)
    ### Sirf Image Upload Karo → Sab Automatic! अब फेक इमेज REAL नहीं बनेगी।
    **Kya karta hai:**
    - 🤖 AI se image ka description nikalta hai
    - 📝 Image mein text padh leta hai (EasyOCR + Tesseract)
    - 🔗 Image type samajhta hai (CLIP) – नए manipulation labels के साथ
    - 🔬 Manipulation check karta hai (ELA + statistics)
    - 🌐 Internet pe real-time search karta hai – **support/contradict check**
    - ✅ Final verdict deta hai – अब सही
    """)
    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### 📸 Image Upload Karo")
            image_input = gr.Image(label="Koi bhi news image, screenshot, ya social media post", type="numpy", height=350)
            analyze_btn = gr.Button("🔍 Analyze Karo!", variant="primary", size="lg")
        with gr.Column(scale=1):
            gr.Markdown("### 📊 Results")
            verdict_out = gr.Textbox(label="🎯 Verdict", interactive=False, text_align="center")
            score_out = gr.Slider(label="Fake News Probability (%)", minimum=0, maximum=100, value=0, interactive=False)
            report_out = gr.Markdown(value="*Image upload karo aur Analyze dabao...*")
    analyze_btn.click(fn=analyze_image_only, inputs=[image_input], outputs=[verdict_out, score_out, report_out])
    gr.Markdown("---\n### 🛠️ Improvements Done:\n- Web search now checks if sources **support or contradict** the claim\n- Search query includes 'hoax' for fake images\n- EasyOCR for better text extraction\n- ELA (Error Level Analysis) for manipulation\n- Adjusted weights & contradict boost\n- Now fake images like 'Eiffel Tower No More' will show **FAKE NEWS**")

print("\n🚀 App launch ho raha hai...")
demo.launch(share=True, debug=False)
```

```python
# ══════════════════════════════════════════════
# SMART FAKE NEWS DETECTOR — CELL 9 UPGRADED
# Sirf Image Upload karo → Automatic sab kuch!
# 1. Image se text read (OCR)
# 2. AI se image samjho (CLIP)
# 3. Internet pe fact check (Web Search)
# 4. AI image manipulation detect karo
# 5. News articles read karo
# 6. Final verdict do!
# ══════════════════════════════════════════════

!pip install -q gradio
!pip install -q duckduckgo-search
!pip install -q pytesseract pillow
!apt-get install -q -y tesseract-ocr
!pip install -q transformers torch
!pip install -q newspaper3k
!pip install -q requests beautifulsoup4
!pip install -q lxml_html_clean

import gradio as gr
import torch
import numpy as np
import requests
from PIL import Image, ImageFilter, ImageEnhance, ImageStat
import pytesseract
import io
import re
import warnings
import time
import json
from bs4 import BeautifulSoup
warnings.filterwarnings('ignore')

try:
    from duckduckgo_search import DDGS
    DDGS_AVAILABLE = True
except:
    DDGS_AVAILABLE = False

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    CLIPProcessor,
    CLIPModel,
    pipeline
)

# ──────────────────────────────────────────────
# MODELS LOAD KARO
# ──────────────────────────────────────────────
print("⏳ Sab models load ho rahe hain...")

# BERT for text fake detection
bert_tok = AutoTokenizer.from_pretrained('distilbert-base-uncased')
try:
    bert_mdl = AutoModelForSequenceClassification.from_pretrained('./bert_best')
    print("✓ Fine-tuned BERT loaded!")
except:
    bert_mdl = AutoModelForSequenceClassification.from_pretrained(
        'distilbert-base-uncased', num_labels=2)
    print("⚠️  Base BERT (fine-tuned nahi mila)")
bert_mdl.eval()

# CLIP for image understanding
print("⏳ CLIP load ho raha hai...")
clip_mdl  = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_proc = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
clip_mdl.eval()

# Image captioning — image ka description banao
print("⏳ Image captioning model load ho raha hai...")
try:
    captioner = pipeline(
        "image-to-text",
        model="Salesforce/blip-image-captioning-base",
        max_new_tokens=50
    )
    CAPTION_AVAILABLE = True
    print("✓ Image captioner ready!")
except Exception as e:
    print(f"⚠️ Captioner nahi chala: {e}")
    CAPTION_AVAILABLE = False

print("✓ Sab models ready!\n")

# ──────────────────────────────────────────────
# CONSTANTS
# ──────────────────────────────────────────────
TRUSTED_SITES = [
    'bbc.com', 'reuters.com', 'apnews.com',
    'theguardian.com', 'nytimes.com',
    'wikipedia.org', 'britannica.com',
    'espn.com', 'cricbuzz.com', 'ndtv.com',
    'thehindu.com', 'indiatoday.in',
    'nationalgeographic.com', 'nasa.gov',
    'who.int', 'un.org', 'iplt20.com',
    'fifa.com', 'uefa.com', 'bcci.tv',
    'snopes.com', 'factcheck.org',
    'politifact.com', 'altnews.in',
    'boomlive.in', 'vishvasnews.com'
]

FAKE_WORDS = [
    'breaking', 'urgent', 'alert', 'leaked',
    'confirmed', 'exposed', 'shocking', 'must',
    'secret', 'exclusive', 'banned', 'censored',
    'bombshell', 'revealed', 'hoax', 'fraud',
    'scam', 'conspiracy', 'hidden truth',
    'they dont want you to know', 'wake up',
    'share before deleted', 'going viral'
]

# ──────────────────────────────────────────────
# HELPER FUNCTIONS
# ──────────────────────────────────────────────

def clean_text(text):
    """Text clean karo"""
    text = str(text).lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def get_bert_score(text):
    """BERT se fake score lo"""
    if not text or len(text.strip()) < 3:
        return 0.5
    inp = bert_tok(
        text, return_tensors='pt',
        padding='max_length', truncation=True, max_length=64
    )
    with torch.no_grad():
        out = bert_mdl(**inp)
    return torch.softmax(out.logits, dim=1)[0][1].item()

def get_image_caption(pil_img):
    """AI se image ka description banao"""
    if not CAPTION_AVAILABLE:
        return None
    try:
        result = captioner(pil_img)
        if result and len(result) > 0:
            return result[0].get('generated_text', '').strip()
    except Exception as e:
        print(f"Caption error: {e}")
    return None

def do_ocr(pil_img):
    """Image se text nikalo (OCR)"""
    try:
        # Image enhance karo better OCR ke liye
        img_gray    = pil_img.convert('L')
        img_sharp   = img_gray.filter(ImageFilter.SHARPEN)
        img_enhance = ImageEnhance.Contrast(img_sharp).enhance(2.0)

        # Multiple PSM modes try karo
        texts = []
        for psm in [6, 3, 11]:
            try:
                t = pytesseract.image_to_string(
                    img_enhance,
                    config=f'--psm {psm}'
                ).strip()
                if t and len(t) > 5:
                    texts.append(t)
            except:
                pass

        if texts:
            # Sabse lamba text return karo
            best = max(texts, key=len)
            # Clean karo
            best = re.sub(r'\n+', ' ', best)
            best = re.sub(r'\s+', ' ', best).strip()
            return best if len(best) > 3 else None
    except Exception as e:
        print(f"OCR error: {e}")
    return None

def get_clip_understanding(pil_img):
    """CLIP se image ka context samjho"""
    try:
        # Alag-alag categories check karo
        categories = [
            # News categories
            "a news photograph",
            "a social media screenshot",
            "a meme or fake image",
            "a real authentic photograph",
            "a digitally manipulated image",
            "a political image",
            "a sports photograph",
            "a disaster or crisis photograph",
            "a celebrity photograph",
            "a protest or rally photograph",
            # Locations
            "a mountain landscape",
            "a city building",
            "a natural disaster",
            "a monument or landmark",
            # Manipulation
            "an AI generated image",
            "a photoshopped image",
        ]

        inputs = clip_proc(
            text=categories,
            images=pil_img,
            return_tensors="pt",
            padding=True
        )
        with torch.no_grad():
            outputs = clip_mdl(**inputs)
            probs = outputs.logits_per_image.softmax(dim=1)[0]

        results = {cat: float(prob) for cat, prob in zip(categories, probs)}
        top3 = sorted(results.items(), key=lambda x: x[1], reverse=True)[:3]
        return results, top3
    except Exception as e:
        print(f"CLIP error: {e}")
        return {}, []

def check_image_manipulation(pil_img):
    """Image manipulation signs dhundo"""
    findings = []
    score    = 0.0

    try:
        # ELA (Error Level Analysis) — simple version
        # JPEG compression artifacts check karo
        img_rgb = pil_img.convert('RGB')

        # Image statistics check
        stat    = ImageStat.Stat(img_rgb)
        r, g, b = stat.stddev[:3]

        # Very uniform image = possibly generated/fake
        if max(r, g, b) < 15:
            findings.append("⚠️ Very uniform colors — possibly AI generated")
            score += 0.3

        # Noise analysis
        img_array = np.array(img_rgb).astype(float)

        # Check for unnatural smoothness in patches
        h, w = img_array.shape[:2]
        patch_size = min(50, h//4, w//4)
        if patch_size > 10:
            patches_std = []
            for i in range(0, h - patch_size, patch_size):
                for j in range(0, w - patch_size, patch_size):
                    patch = img_array[i:i+patch_size, j:j+patch_size]
                    patches_std.append(np.std(patch))

            avg_std = np.mean(patches_std)
            if avg_std < 8:
                findings.append("⚠️ Unusually smooth texture — possible AI/CGI image")
                score += 0.25

        # Resolution check
        width, height = pil_img.size
        if width < 200 or height < 200:
            findings.append("⚠️ Very low resolution image")
            score += 0.1

        # Aspect ratio check
        ratio = width / height
        if ratio > 4 or ratio < 0.25:
            findings.append("⚠️ Unusual aspect ratio")
            score += 0.1

    except Exception as e:
        print(f"Manipulation check error: {e}")

    return findings, min(score, 1.0)

def web_search_and_read(query, max_results=8):
    """Internet pe search karo aur articles padho"""
    trusted_sources = []
    all_sources     = []
    article_content = []
    fact_checks     = []

    if not DDGS_AVAILABLE:
        return trusted_sources, all_sources, article_content, fact_checks

    try:
        with DDGS() as ddgs:
            results = list(ddgs.text(query, max_results=max_results))

        for r in results:
            url    = r.get('href', '')
            title  = r.get('title', '')
            body   = r.get('body', '')[:200]

            source_info = {
                'url': url, 'title': title, 'body': body,
                'trusted': any(s in url for s in TRUSTED_SITES)
            }

            all_sources.append(source_info)
            if source_info['trusted']:
                trusted_sources.append(source_info)

            # Fact check sites
            if any(fc in url for fc in ['snopes', 'factcheck', 'politifact', 'boomlive', 'altnews', 'vishvasnews']):
                fact_checks.append(source_info)

            # Article content read karo (sirf trusted sites ke liye)
            if source_info['trusted'] and len(article_content) < 3:
                try:
                    headers = {
                        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
                    }
                    resp = requests.get(url, timeout=5, headers=headers)
                    if resp.status_code == 200:
                        soup = BeautifulSoup(resp.content, 'html.parser')
                        # Main content nikalo
                        for tag in soup(['script', 'style', 'nav', 'header', 'footer', 'ads']):
                            tag.decompose()
                        paragraphs = soup.find_all('p')
                        content    = ' '.join([p.get_text() for p in paragraphs[:5]])
                        content    = re.sub(r'\s+', ' ', content).strip()
                        if content and len(content) > 100:
                            article_content.append({
                                'url': url, 'title': title,
                                'content': content[:500]
                            })
                except Exception as e:
                    pass  # Article padh nahi paya — chalega

    except Exception as e:
        print(f"Search error: {e}")

    return trusted_sources, all_sources, article_content, fact_checks

def check_claim_vs_image(image_caption, ocr_text, clip_top3):
    """Image aur text mein match check karo"""
    issues   = []
    mismatch = 0.0

    if not image_caption:
        return issues, 0.3

    caption_lower = image_caption.lower()

    # CLIP top result check
    if clip_top3:
        top_label = clip_top3[0][0].lower()
        if 'manipulated' in top_label or 'photoshop' in top_label or 'ai generated' in top_label:
            issues.append(f"🚨 AI/CLIP detects: '{clip_top3[0][0]}' (score: {clip_top3[0][1]:.1%})")
            mismatch += 0.5
        elif 'fake' in top_label or 'meme' in top_label:
            issues.append(f"⚠️ Image looks like: '{clip_top3[0][0]}'")
            mismatch += 0.3

    # OCR text check
    if ocr_text:
        ocr_lower = ocr_text.lower()
        fake_found = [w for w in FAKE_WORDS if w in ocr_lower]
        if fake_found:
            issues.append(f"⚠️ Suspicious words in image text: {fake_found}")
            mismatch += 0.2 * len(fake_found)

    return issues, min(mismatch, 1.0)

# ══════════════════════════════════════════════
# MAIN DETECTION FUNCTION
# ══════════════════════════════════════════════

def analyze_image_only(image):
    """
    Sirf image do — sab kuch automatically check hoga!
    """

    if image is None:
        return (
            "⚠️ Koi image upload nahi ki!",
            0,
            "Pehle ek image upload karo."
        )

    report  = []
    scores  = {}
    all_findings = []
    start_time   = time.time()

    # PIL image banao
    if not isinstance(image, Image.Image):
        pil_img = Image.fromarray(image)
    else:
        pil_img = image

    report.append("# 🔍 Smart Fake News Analysis\n")
    report.append(f"Image size: {pil_img.size[0]}×{pil_img.size[1]} pixels\n")
    report.append("---\n")

    # ══════════════════════
    # STEP 1: IMAGE CAPTION
    # AI se samjho — image mein kya hai?
    # ══════════════════════
    report.append("## 👁️ Step 1: Image ko Samjho (AI Caption)")
    image_caption = get_image_caption(pil_img)

    if image_caption:
        report.append(f"- 🤖 AI Caption: **\"{image_caption}\"**")
        all_findings.append(f"Image shows: {image_caption}")
        search_query = image_caption  # Ye query use karenge search mein
    else:
        report.append("- ⚠️ Caption generate nahi hua")
        search_query = "news fact check"

    report.append("")

    # ══════════════════════
    # STEP 2: OCR — IMAGE MEIN TEXT NIKALO
    # ══════════════════════
    report.append("## 📝 Step 2: Image Mein Text Dhundho (OCR)")
    ocr_text = do_ocr(pil_img)

    if ocr_text and len(ocr_text) > 5:
        report.append(f"- ✓ Image mein text mila: **\"{ocr_text[:150]}\"**")
        all_findings.append(f"OCR text: {ocr_text[:100]}")

        # OCR text ko bhi search query mein add karo
        # Clean karo pehle
        ocr_clean = re.sub(r'[^a-zA-Z0-9\s]', ' ', ocr_text)
        ocr_clean = re.sub(r'\s+', ' ', ocr_clean).strip()
        if len(ocr_clean) > 10:
            search_query = ocr_clean[:100]  # OCR text zyada accurate hoga

        # BERT se text analyze karo
        bert_score = get_bert_score(clean_text(ocr_text[:200]))
        scores['bert_text'] = bert_score
        report.append(f"- 🤖 BERT Text Score: **{bert_score:.1%}** fake probability")

        # Fake words check
        fake_found = [w for w in FAKE_WORDS if w in ocr_text.lower()]
        if fake_found:
            report.append(f"- 🚨 Suspicious words mile: **{fake_found}**")
            all_findings.append(f"Suspicious words: {fake_found}")
        else:
            report.append("- ✅ Koi suspicious words nahi")
    else:
        report.append("- ℹ️ Image mein koi readable text nahi mila")
        scores['bert_text'] = 0.4  # Default neutral score

    report.append("")

    # ══════════════════════
    # STEP 3: CLIP — IMAGE TYPE SAMJHO
    # ══════════════════════
    report.append("## 🔗 Step 3: Image Type Analysis (CLIP AI)")
    clip_all, clip_top3 = get_clip_understanding(pil_img)
    manipulation_score  = 0.0

    if clip_top3:
        report.append("- Top matches:")
        for label, prob in clip_top3:
            bar = "█" * int(prob * 30)
            report.append(f"  - `{label}`: **{prob:.1%}** {bar}")

        top_label = clip_top3[0][0].lower()
        top_prob  = clip_top3[0][1]

        if any(w in top_label for w in ['manipulated', 'photoshop', 'ai generated', 'fake', 'meme']):
            manipulation_score = top_prob * 0.8
            report.append(f"\n- 🚨 **WARNING: Image suspicious lagti hai!**")
            all_findings.append(f"CLIP says: '{clip_top3[0][0]}'")
        else:
            report.append(f"\n- ✅ Image type: '{clip_top3[0][0]}'")

        scores['clip_manipulation'] = manipulation_score
    report.append("")

    # ══════════════════════
    # STEP 4: IMAGE MANIPULATION CHECK
    # ══════════════════════
    report.append("## 🔬 Step 4: Image Manipulation Check")
    manip_findings, manip_score = check_image_manipulation(pil_img)

    if manip_findings:
        for f in manip_findings:
            report.append(f"- {f}")
            all_findings.append(f)
        scores['manipulation'] = manip_score
        report.append(f"- Manipulation Score: **{manip_score:.1%}**")
    else:
        report.append("- ✅ No obvious manipulation detected")
        scores['manipulation'] = 0.1

    report.append("")

    # ══════════════════════
    # STEP 5: INTERNET FACT CHECK
    # Real-time web search + article reading!
    # ══════════════════════
    report.append("## 🌐 Step 5: Internet Fact Check (Real-time)")
    report.append(f"- Searching for: **\"{search_query[:80]}\"**\n")

    trusted_src, all_src, articles, fact_checks = web_search_and_read(search_query)

    n_trusted    = len(trusted_src)
    n_total      = len(all_src)
    n_factchecks = len(fact_checks)

    # Web score calculate
    if   n_trusted >= 3: web_score = 0.05
    elif n_trusted == 2: web_score = 0.15
    elif n_trusted == 1: web_score = 0.30
    elif n_total   >= 4: web_score = 0.55
    elif n_total   >= 1: web_score = 0.70
    else:                web_score = 0.90

    # Fact checkers mila toh aur better
    if n_factchecks > 0:
        web_score = max(0.05, web_score - 0.1)

    scores['web'] = web_score

    report.append(f"- 🔎 Total results: **{n_total}**")
    report.append(f"- ✅ Trusted sources: **{n_trusted}**")
    if n_factchecks > 0:
        report.append(f"- 🏆 Fact-checkers: **{n_factchecks}**")

    # Trusted sources dikhao
    if trusted_src:
        report.append("\n### ✅ Trusted Sources:")
        for s in trusted_src[:4]:
            report.append(f"- 🔗 [{s['title'][:60]}]({s['url']})")
            if s.get('body'):
                report.append(f"  > {s['body'][:120]}...")

    # Fact checks
    if fact_checks:
        report.append("\n### 🏆 Fact-Check Results:")
        for fc in fact_checks[:2]:
            report.append(f"- 🔗 [{fc['title'][:60]}]({fc['url']})")
            if fc.get('body'):
                report.append(f"  > {fc['body'][:150]}...")
        all_findings.append(f"Fact-check sites found: {[fc['url'] for fc in fact_checks[:2]]}")

    # Article content
    if articles:
        report.append("\n### 📰 Article Excerpts (Auto-Read):")
        for art in articles[:2]:
            report.append(f"\n**{art['title'][:60]}**")
            report.append(f"> {art['content'][:300]}...")
            report.append(f"Source: {art['url'][:60]}")

    if not all_src:
        report.append("\n- ❌ **Koi internet result nahi mila — bahut suspicious!**")
        all_findings.append("No internet results found")

    report.append("")

    # ══════════════════════
    # STEP 6: CAPTION vs INTERNET MATCH
    # Image mein jo hai aur internet pe jo hai — match karta hai?
    # ══════════════════════
    if image_caption and all_src:
        report.append("## 🔄 Step 6: Image vs Internet Match")
        caption_words = set(clean_text(image_caption).split())

        match_count = 0
        for src in all_src[:5]:
            src_text  = clean_text(src.get('title', '') + ' ' + src.get('body', ''))
            src_words = set(src_text.split())
            overlap   = caption_words & src_words
            if len(overlap) >= 2:
                match_count += 1

        match_ratio = match_count / min(5, len(all_src))
        scores['caption_web_match'] = 1.0 - match_ratio

        if match_ratio > 0.6:
            report.append(f"- ✅ Image content matches internet ({match_count}/{min(5, len(all_src))} sources agree)")
        elif match_ratio > 0.3:
            report.append(f"- ⚠️ Partial match ({match_count}/{min(5, len(all_src))} sources)")
            all_findings.append("Partial image-internet match")
        else:
            report.append(f"- 🚨 **Low match! Image content aur internet news mein fark!**")
            all_findings.append("Image content doesn't match internet news")
    else:
        scores['caption_web_match'] = 0.4

    report.append("")

    # ══════════════════════
    # STEP 7: FINAL COMBINED SCORE
    # ══════════════════════
    report.append("## 📊 Final Score Calculation")

    # Weights
    w_bert    = 0.15
    w_clip    = 0.15
    w_manip   = 0.15
    w_web     = 0.35
    w_caption = 0.20

    final = (
        scores.get('bert_text', 0.4)          * w_bert    +
        scores.get('clip_manipulation', 0.1)  * w_clip    +
        scores.get('manipulation', 0.1)       * w_manip   +
        scores.get('web', 0.5)                * w_web     +
        scores.get('caption_web_match', 0.4)  * w_caption
    )

    report.append(f"| Component | Score | Weight |")
    report.append(f"|-----------|-------|--------|")
    report.append(f"| BERT text | {scores.get('bert_text', 0.4):.1%} | 15% |")
    report.append(f"| CLIP manipulation | {scores.get('clip_manipulation', 0.1):.1%} | 15% |")
    report.append(f"| Image analysis | {scores.get('manipulation', 0.1):.1%} | 15% |")
    report.append(f"| Web fact-check | {scores.get('web', 0.5):.1%} | 35% |")
    report.append(f"| Caption match | {scores.get('caption_web_match', 0.4):.1%} | 20% |")
    report.append(f"| **FINAL** | **{final:.1%}** | 100% |")
    report.append("")

    # ══════════════════════
    # VERDICT
    # ══════════════════════
    elapsed = time.time() - start_time

    if   final >= 0.75:
        verdict = "🔴 FAKE NEWS"
        conf    = "Bahut zyada chance hai ye FAKE hai!"
    elif final >= 0.60:
        verdict = "🟠 Likely FAKE"
        conf    = "Suspicious — zaroor verify karo"
    elif final >= 0.45:
        verdict = "🟡 Uncertain"
        conf    = "Confirm nahi ho sakta — manually check karo"
    elif final >= 0.25:
        verdict = "🟡 Likely REAL"
        conf    = "Probably real news hai"
    else:
        verdict = "🟢 REAL NEWS"
        conf    = "Sources verified — real lagta hai"

    report.append("---")
    report.append(f"# {verdict}")
    report.append(f"**{conf}**")
    report.append(f"\n**Final Score: {final:.1%}** | Analysis time: {elapsed:.1f}s")

    if all_findings:
        report.append("\n## 🔍 Key Findings:")
        for f in all_findings[:8]:
            report.append(f"- {f}")

    report.append("\n---")
    report.append("*Analysis: AI Caption + OCR + CLIP + Manipulation Detection + Web Search + Article Reading*")

    return verdict, int(final * 100), "\n".join(report)


# ══════════════════════════════════════════════
# GRADIO UI
# ══════════════════════════════════════════════

with gr.Blocks(title="Smart Fake News Detector", theme=gr.themes.Soft()) as demo:

    gr.Markdown("""
    # 🔍 Smart Fake News Detector
    ### Sirf Image Upload Karo → Sab Automatic!
    **Kya karta hai:**
    - 🤖 AI se image ka description nikalta hai
    - 📝 Image mein text padh leta hai (OCR)
    - 🔗 Image type samajhta hai (CLIP)
    - 🔬 Manipulation check karta hai
    - 🌐 Internet pe real-time search karta hai
    - 📰 News articles khud padh leta hai
    - ✅ Final verdict deta hai!
    """)

    with gr.Row():
        with gr.Column(scale=1):
            gr.Markdown("### 📸 Image Upload Karo")
            image_input = gr.Image(
                label="Koi bhi news image, screenshot, ya social media post",
                type="numpy",
                height=350
            )
            analyze_btn = gr.Button(
                "🔍 Analyze Karo!",
                variant="primary",
                size="lg"
            )
            gr.Markdown("""
            **Kya upload kar sakte ho:**
            - 📰 News article screenshots
            - 📱 WhatsApp forwards
            - 🐦 Twitter/Instagram posts
            - 🖼️ Any image with news claim
            - 📸 Photos with captions
            """)

        with gr.Column(scale=1):
            gr.Markdown("### 📊 Results")
            verdict_out = gr.Textbox(
                label="🎯 Verdict",
                interactive=False,
                text_align="center"
            )
            score_out = gr.Slider(
                label="Fake News Probability (%)",
                minimum=0, maximum=100, value=0,
                interactive=False,
                info="0% = Definitely Real | 100% = Definitely Fake"
            )
            report_out = gr.Markdown(
                value="*Image upload karo aur Analyze dabao...*"
            )

    analyze_btn.click(
        fn=analyze_image_only,
        inputs=[image_input],
        outputs=[verdict_out, score_out, report_out]
    )

    gr.Markdown("""
    ---
    ### 🛠️ Kaise Kaam Karta Hai?

    | Step | Method | Kya karta hai |
    |------|--------|---------------|
    | 1 | AI Caption (BLIP) | Image mein kya hai ye samajhta hai |
    | 2 | OCR (Tesseract) | Image ka text padh leta hai |
    | 3 | BERT | Text ka fake score nikalta hai |
    | 4 | CLIP (OpenAI) | Image ka type identify karta hai |
    | 5 | Manipulation Check | AI/Photoshop signs dhundhta hai |
    | 6 | Web Search | Internet pe real-time search karta hai |
    | 7 | Article Reader | News articles khud padh ke verify karta hai |
    | 8 | Ensemble | Sab milake final verdict deta hai |

    ### 🎯 Score Guide
    - 🟢 **0-25%** = Real News
    - 🟡 **25-45%** = Probably Real
    - 🟡 **45-60%** = Uncertain — manually verify karo
    - 🟠 **60-75%** = Likely Fake
    - 🔴 **75-100%** = Fake News!

    *Dataset: FakeNewsNet | Paper: Al-Tai et al. 2023*
    """)

print("\n🚀 App launch ho raha hai...")
demo.launch(share=True, debug=False)
```
